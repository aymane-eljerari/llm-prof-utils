```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Placeholder path to the CSV file
csv_file_path = "/app/performance_results/F16_none.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Function to plot the histogram for each tensor
def plot_histograms(df):
    for index, row in df.iterrows():
        tensor_name = row['Tensor Name']
        min_value = row['Min Weight']
        max_value = row['Max Weight']
        bin_values = row[['Bin 0', 'Bin 1', 'Bin 2', 'Bin 3', 'Bin 4', 
                          'Bin 5', 'Bin 6', 'Bin 7', 'Bin 8', 'Bin 9']].values
        
        # Create 10 bins between the min and max values
        bins = np.linspace(min_value, max_value, 11)
        
        # Plot the histogram
        plt.figure(figsize=(8, 6))
        plt.bar(bins[:-1], bin_values, width=np.diff(bins), edgecolor="black", align="edge")
        
        # Set title and labels
        plt.title(f'{tensor_name} Histogram')
        plt.xlabel(f'Value Range: [{min_value:.6f}, {max_value:.6f}]')
        plt.ylabel('Frequency')
        
        # Show the x-axis values with min and max explicitly visible
        plt.xticks(ticks=bins, labels=[f'{edge:.6f}' for edge in bins], rotation=45)
        
        # Display the plot
        plt.tight_layout()
        plt.show()

# Call the function to plot the histograms
plot_histograms(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import matplotlib.colors as mcolors

# Define base colors for each quant type
base_colors = {
    "F32": "#0000FF",  # Blue
    "F16": "#FF0000",  # Red
    "Q8_0": "#00FF00", # Green
    "Q5_1": "#800080", # Purple
    "Q4_1": "#FFA500"  # Orange
}

def generate_color_gradient(base_color, n_shades=3):
    """Generate a gradient of colors from a base color."""
    base_rgb = mcolors.hex2color(base_color)
    # Generate a list of shades from the base color
    return [mcolors.to_hex(np.array(base_rgb) * (1 - i / (n_shades - 1)) + np.array([1, 1, 1]) * (i / (n_shades - 1))) for i in range(n_shades)]

def get_color_gradients():
    """Return a dictionary with color gradients for each quant type."""
    return {quant: generate_color_gradient(color) for quant, color in base_colors.items()}

def load_all_data():
    all_data = {}
    for quant in quants:
        for splitmode in split_modes:
            file_path = f"performance_results/{quant}_{splitmode}.csv"
            # Skip the specific file
            if file_path == "performance_results/F32_none.csv":
                continue
            if os.path.isfile(file_path):
                df = pd.read_csv(file_path)
                all_data[(quant, splitmode)] = df
            else:
                print(f"File not found: {file_path}")
    return all_data

def filter_data(df, q, sm, y_col, pp_values, tg_values, b_values):
    filtered_df = df[(df['PP'].isin(pp_values)) & 
                      (df['TG'].isin(tg_values)) & 
                      (df['B'].isin(b_values))]
    filtered_df = filtered_df[[y_col]]
    return filtered_df

def aggregate_data(y_col, pp_values, tg_values, b_values):
    all_data = load_all_data()
    aggregated_data = pd.DataFrame()
    
    for (quant, splitmode), df in all_data.items():
        filtered_df = filter_data(df, quant, splitmode, y_col, pp_values, tg_values, b_values)
        filtered_df['quant'] = quant
        filtered_df['split_mode'] = splitmode
        aggregated_data = pd.concat([aggregated_data, filtered_df], ignore_index=True)
    
    aggregated_data['pp_values'] = str(pp_values)
    aggregated_data['tg_values'] = str(tg_values)
    aggregated_data['b_values'] = str(b_values)
    
    return aggregated_data

def generate_bar_plot(df):
    required_columns = ['quant', 'split_mode']
    if len(df.columns) < 3:
        raise ValueError("The DataFrame must have at least three columns.")

    value_col = df.columns[0]
    df['label'] = df['quant'] + '_' + df['split_mode']
    
    # Generate color gradients
    color_gradients = get_color_gradients()
    
    plt.figure(figsize=(12, 5))
    
    # Create a color map for each quant type
    for quant, group in df.groupby('quant'):
        colors = color_gradients.get(quant, ["skyblue", "skyblue", "skyblue"])
        print(colors)
        for split_mode in split_modes:
            sub_group = group[group['split_mode'] == split_mode]
            color = colors[split_modes.index(split_mode)]
            plt.bar(
                sub_group['label'],
                sub_group[value_col],
                color=color,
                edgecolor='black',
                label=f"{quant} - {split_mode}",
                width=0.8  # Adjust width to avoid overlapping
            )
    
    plt.xlabel('Quant and Split Mode')
    plt.ylabel(value_col)
    plt.title(f'Bar Plot of {value_col} by Quant and Split Mode')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.show()

def main():
    df = aggregate_data("S_PP t/s", [512], [128], [8])
    print(df)
    generate_bar_plot(df)

if __name__ == "__main__":
    main()
```
